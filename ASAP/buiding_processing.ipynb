```python
X=[]
Y=[]
Z=[]
intensity=[]
flag_byte=[]
raw_classification=[]
scan_angle_rank=[]
user_data=[]
pt_src_id=[]
gps_time=[]
```

```python
for n in np.arange(len(point)):
    X.append(point[n][0][0])
    Y.append(point[n][0][1])
    Z.append(point[n][0][2])
    intensity.append(point[n][0][3])
    flag_byte.append(point[n][0][4])
    raw_classification.append(point[n][0][5])
    scan_angle_rank.append(point[n][0][6])
    user_data.append(point[n][0][7])
    pt_src_id.append(point[n][0][8])
    gps_time.append(point[n][0][9])
```

```python
import pandas as pd
```

```python
buiding = pd.DataFrame(data={"X":X,"Y":Y,"Z":Z,"intensity":intensity,"flag_byte":flag_byte,
                            "raw_classification":raw_classification,"scan_angle_rank":scan_angle_rank,
                            "user_data":user_data,"pt_src_id":pt_src_id,"gps_time":gps_time})
```

```python
buiding
```

```python
buiding.to_csv('buiding.csv')
```
